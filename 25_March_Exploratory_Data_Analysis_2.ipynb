##Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

##Ans:--

###Here's the Python code to create a histogram of flight prices in the dataset:

```python
import matplotlib.pyplot as plt

# Load the dataset from the Google Sheets link
url = 'https://docs.google.com/spreadsheets/d/10N5ItjuFCg7iJx-_eXklZENr9mhf2l57/export?format=csv'
flight_df = pd.read_csv(url)

# Create a histogram of flight prices
plt.hist(flight_df['Price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

###The histogram shows that the distribution of flight prices in the dataset is right-skewed, which means that there are more flights with lower prices and fewer flights with higher prices. The majority of the flights have prices between 5,000 to 15,000

##Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

##Ans:--

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the Google Sheets link
url = 'https://docs.google.com/spreadsheets/d/10N5ItjuFCg7iJx-_eXklZENr9mhf2l57/export?format=csv'
flight_df = pd.read_csv(url)

# Set the size of the plot
plt.figure(figsize=(15, 7))

# Create a boxplot to compare the prices of different airlines
sns.boxplot(x='Airline', y='Price', data=flight_df)
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Price Distribution by Airline')

# Rotate the x-tick labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

##Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

##Ans:---

###To explore the relationship between the size of an app and its rating, we can create a scatter plot with the "Size" column on the x-axis and the "Rating" column on the y-axis. This will allow us to see if there is any correlation between these two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
playstore_df = pd.read_csv("/content/google_cleaned.csv")

# Convert "Varies with device" to NaN
# playstore_df["Size"].replace("Varies with device", pd.np.nan, inplace=True)

# Convert "Size" column to float
playstore_df["Size"] = pd.to_numeric(playstore_df["Size"])

# Drop rows with missing values in "Size" column
playstore_df.dropna(subset=["Size"], inplace=True)

# Create scatter plot
plt.scatter(playstore_df["Size"], playstore_df["Rating"])
plt.xlabel("Size (in MB)")
plt.ylabel("Rating")
plt.title("Relationship between App Size and Rating")
plt.show()
```

###The scatter plot will show us if there is a positive or negative correlation between the size of an app and its rating. If there is a positive correlation, we can expect to see a general trend of higher ratings for larger apps. If there is a negative correlation, we can expect to see a general trend of lower ratings for larger apps. If there is no correlation, we can expect to see a random distribution of ratings across app sizes.

###So in this scatter plot we can see that low size app are having higher ratings

##Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

##Ans:---

###The type of app can affect its price. Paid apps generally have a higher price than free apps. Additionally, some app categories such as education or business may have a higher price compared to entertainment or social networking apps. The analysis of average prices by app type can help to identify any such patterns or trends.
###To create a bar chart to compare average prices by app type, we can group the dataset by the App Type column and calculate the mean price for each group. Then, we can plot the results using a bar chart.

###Here's the Python code to do this:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
df = pd.read_csv("/content/google_cleaned.csv")

# Group the dataset by App Type and calculate the mean price for each group
grouped = df.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.bar(grouped.index, grouped.values)
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

###bar chart that shows the average price for each app type. The x-axis represents the app type (Free or Paid) and the y-axis represents the average price.

##Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

##Ans:----




```python
import pandas as pd

# read the dataset
df = pd.read_csv("/content/google_cleaned.csv")

# create a frequency table of the top 10 most popular apps by number of installs
top_10_apps = df[['App', 'Installs']].groupby('App').sum().sort_values(by='Installs', ascending=False).head(10)

# beautify the output
top_10_apps.style.format({'Installs': '{:,}'}).bar(subset=['Installs'], color='lighred', vmax=top_10_apps['Installs'].max()/2)
```
